```python
# test connection

# requests.get(link)

link = 'https://id.wikipedia.org/wiki/Provinsi_di_Indonesia'
# link = 'https://www.bukalapak.com/products?from=omnisearch&from_keyword_history=false&search%5Bkeywords%5D=sepeda%20lipat&search_source=omnisearch_keyword&source=navbar'
try:
    response = requests.get(link)
    if response.status_code == 200:
        print('requests http terhadap link berhasil')
    else:
        print('requests http terhadap link berhasil tetapi perlu hak akses')
except:
    print('requests http terhadap link tidak berhasil')
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})
table
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})

header = []
data = []

for th in table.find_all('th'):
    header.append(th.get_text(strip=True))

# header

index_kode = header.index('Kode')  # Menemukan posisi 'Kode'
header_modified = header[:index_kode + 1] + ['ID',] + header[index_kode + 1:]

index_t2 = header_modified.index('Tingkat II')
index_t3 = header_modified.index('#[3]')
header_modified[index_t2] = 'Kab'
header_modified[index_t3] = 'Kota'

for row in table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) > 0:
        data.append([col.get_text(strip=True) for col in cols])

df = pd.DataFrame(data, columns=header_modified[:len(data[0])])
df['Hari jadi'] = df['Hari jadi'].str[:-4]
mask = df['Kab'].str.contains(r'\[.*\]')
df.loc[mask, 'Kab'] = df.loc[mask, 'Kab'].str[:-3]
mask = df['Kota'].str.contains(r'\[.*\]')
df.loc[mask, 'Kota'] = df.loc[mask, 'Kota'].str[:-3]

df
```
